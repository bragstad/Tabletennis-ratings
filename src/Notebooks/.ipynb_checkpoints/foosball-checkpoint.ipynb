```python
# Calculate average goals per game
players = list(f_players.keys())
forward_avg_goals = [f_players[player]['forward_goals'] / f_players[player]['games_played'] if f_players[player]['games_played'] > 0 else 0 for player in players]
midfield_avg_goals = [f_players[player]['midfield_goals'] / f_players[player]['games_played'] if f_players[player]['games_played'] > 0 else 0 for player in players]
defence_avg_goals = [f_players[player]['defence_goals'] / f_players[player]['games_played'] if f_players[player]['games_played'] > 0 else 0 for player in players]
goalie_avg_goals = [f_players[player]['goalie_goals'] / f_players[player]['games_played'] if f_players[player]['games_played'] > 0 else 0 for player in players]

# Calculate total average goals per game for sorting
total_avg_goals = [f_players[player]['goals'] / f_players[player]['games_played'] if f_players[player]['games_played'] > 0 else 0 for player in players]

# Sort players based on total average goals per game
sorted_indices = np.argsort(total_avg_goals)[::-1]  # Sort in descending order
players_sorted = [players[i] for i in sorted_indices]
forward_avg_goals_sorted = [forward_avg_goals[i] for i in sorted_indices]
midfield_avg_goals_sorted = [midfield_avg_goals[i] for i in sorted_indices]
defence_avg_goals_sorted = [defence_avg_goals[i] for i in sorted_indices]
goalie_avg_goals_sorted = [goalie_avg_goals[i] for i in sorted_indices]

# Number of bars
n = len(players)

# Position of bars on x-axis
ind = np.arange(n)

# Width of a bar
width = 0.35       

# Plotting
fig, ax = plt.subplots()
p1 = ax.bar(ind, forward_avg_goals_sorted, width, label='Forward Goals')
p2 = ax.bar(ind, midfield_avg_goals_sorted, width, bottom=forward_avg_goals_sorted, label='Midfield Goals')
p3 = ax.bar(ind, defence_avg_goals_sorted, width, bottom=np.array(forward_avg_goals_sorted) + np.array(midfield_avg_goals_sorted), label='Defence Goals')
p4 = ax.bar(ind, goalie_avg_goals_sorted, width, bottom=np.array(forward_avg_goals_sorted) + np.array(midfield_avg_goals_sorted) + np.array(defence_avg_goals_sorted), label='Goalie Goals')

ax.axhline(0, color='grey', linewidth=0.8)
ax.set_xlabel('Players')
ax.set_ylabel('Average Goals per Game')
ax.set_title('Average Goals per Game by Player and Position')
ax.set_xticks(ind)
ax.set_xticklabels(players_sorted, rotation=45, ha='right')
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Assuming foosball_goals is already defined
df = pd.DataFrame(foosball_goals, columns=['goal_id', 'position', 'team', 'human_player'])

# Group by position and player, then count goals
goal_summary = df.groupby(['position', 'human_player']).size().unstack(fill_value=0)

# Calculate total goals per position and sort the DataFrame based on this
total_goals_per_position = goal_summary.sum(axis=1).sort_values(ascending=False)
sorted_goal_summary = goal_summary.loc[total_goals_per_position.index]

# Calculate total goals by player for legend
total_goals_by_player = df.groupby('human_player').size()

# Plot grouped bar chart with sorted positions
ax = sorted_goal_summary.plot(kind='bar', stacked=False, figsize=(12, 8), colormap='viridis')
plt.xlabel('Position')
plt.ylabel('Goals Scored')
plt.title('Goals Scored by Position and Player (Sorted by Position Total)')

# Customize legend to include total goals by player
handles, labels = ax.get_legend_handles_labels()
new_labels = [f"{label} ({total_goals_by_player[label]} goals)" for label in labels]

# Adding total goals per position to the legend
position_labels = [f"{pos} ({total_goals_per_position[pos]} goals)" for pos in total_goals_per_position.index]
plt.legend(handles, new_labels, title='Player', bbox_to_anchor=(1.05, 1))

# Display total goals per position in the legend
for pos_label in position_labels:
    plt.text(1.05, 0.95 - 0.05 * position_labels.index(pos_label), pos_label, transform=ax.transAxes)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
